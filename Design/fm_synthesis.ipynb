```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig


def fm_synth(f_c: float, ratio: float, b: float, n: np.arange) -> np.arange:
    # f_c: carrier frequency
    # ratio: multiplier for
    # b: scaler
    modulator = b*np.sin(ratio*f_c*n)
    return np.sin(f_c*n + modulator)


def plot_f(fc: float, ratio: float, b: float):
    n = np.arange(2**14)
    x = fm_synth(fc, ratio, b, n)
    _, ax = plt.subplots(2)
    ax[0].plot(n[:128], x[:128])
    f, X = sig.welch(x, nperseg=1024)
    ax[1].plot(f, X)

```

```python
fs = 48000
f = 880

plot_f(2*np.pi*f/fs, 1, 1)
```

```python
fs = 48000
f = 2000

plot_f(2*np.pi*f/fs, 2, 1.5)
```

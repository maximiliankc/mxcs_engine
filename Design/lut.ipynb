```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

fs = 48000
points = 4096

t = np.arange(points)

lut = np.sin((t/points)*2*np.pi)

_, ax = plt.subplots()

ax.plot(t/points, lut, marker='.')
ax.grid()

N = 4
n = 1
for sample in lut:
    print(f"{sample:.18}", end=', ')
    if n % N == 0:
        print("")
    n += 1
```

```python
f = 501
n = np.arange(2**6)

phases = f/fs*n
out = np.sin(2*np.pi*phases)
indexes = [int(phase*points)%points for phase in phases]
print(indexes)
out_lut = [lut[idx] for idx in indexes]

_, ax = plt.subplots(2)
ax[0].plot(n/fs, out)
ax[0].plot(n/fs, out_lut)
ax[0].grid()

ax[1].plot(out - np.array(out_lut))

```
